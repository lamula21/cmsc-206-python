```python
# Jose A Valdivia Rojas
# CMSC 206
# Assignment 5_1
# NOTE: WHEN RUN THIS PROGRAM, MINIMIZE YOUR TEXT EDITOR or IDE SO THAT YOU ABLE TO CHOOSE YOUR FILE IN A NEW WINDOWs
# YOU MUST CHOOSE A FILE, OTHERWISE THE WINDOW WILL BE OPENED FOREVER. TO STOP IT, INTERRUPT THE PROGRAM


# Variables
state_and_region_dataList = []

# Create FileChooser in Python version
from tkinter import filedialog
def openFile():
    print("Choose a csv file: ")
    user_file = filedialog.askopenfilename()
    return user_file    # user_file is a string that has a PATH of the file



# Process the file
import csv  # Import csv to process a csv file

#Get the path file from user by using the function openFile() created above
user_file = openFile()

# Then open file with the PATH from user and valid the file
while True:
    try:
        with open(user_file,'r') as csv_file:
            csv_reader_object = csv.reader(csv_file)    # Create an object that reads the csv file
            data_list = [ lines for lines in csv_reader_object]  # Read the data from csv file to a new a list of lists
            break
    except:
        print("Please choose a valid file.")
        user_file = openFile()
    
print(data_list)

# Get the user input
user_state = input("Enter a STATE, or NO to skip")
user_region = input("Enter a REGION, or NO to skip")

user_state = ( user_state.title() ).replace(" ", "_")     # Capitalize first letter of each word from user input, then replace "whitespace" to "_"        
user_region = ( user_region.title() ).replace(" ", "_")   # Capitalize first letter of each word from user input, then replace "whitespace" with "_"

# User choices
print("What would you like to know from the STATE or REGION? Choose 2 options")
print("1) Population")
print("2) GDP")
print("3) Personal Income")
print("4) Subsidies")
print("5) Comp of Emp")
print("6) Tax on Prod/Imp")
print("")

# Validate user choices
while True:
    try:
        user_choice_1 = int(input("Enter a choice from 1 to 6: "))
        if user_choice_1 >= 1 or user_choice_1 <= 6:
            break
        else:
            print("Wrong choice. Enter a number from 1 to 6")
    except ValueError:
        print("Please re-enter a valid number from 1 to 6")

while True:
    try:
        user_choice_2 = int(input("Enter another choice from 1 to 6: "))
        if user_choice_2 >= 1 or user_choice_2 <= 6:
            break
        else:
            print("Wrong choice. Enter a number from 1 to 6")
    except ValueError:
        print("Please re-enter a valid number from 1 to 6")



#  Store relevant information, according to STATE and REGION chosen from the user, in a list of lists
for eachLine in data_list:
    try:
        if eachLine.index(user_state) >= 0:
            state_and_region_dataList.append(eachLine)
    except:
        continue

for eachLine in data_list:
    try:
        if eachLine.index(user_region) >= 0:
            state_and_region_dataList.append(eachLine) 
    except:        
        continue
        

# Sort the dataList
print("")
state_and_region_dataList.sort()


# No duplicate lines in a list of lists
unique_dataList = []
for eachLine in state_and_region_dataList:
    if eachLine not in unique_dataList:         # in â€“ membership operator to check if the line is present in unique_dataList
        unique_dataList.append(eachLine)



# Header List
header = ["STATE", "REGION", "POPULATION", "GDP", "PERSONAL INCOME", "SUBSIDIES", "COMP OF EMP", "TAX ON PROD/IMP"]
    
# Display header and format it, >15 means right justified with 15 spaces
print ("{:>15} {:>15} {:>15} {:>15}".format(header[0], header[1], header[user_choice_1 + 1], header[user_choice_2 + 1]))

# Query revelant information and format it
for eachLine in unique_dataList:
    print ("{:>15} {:>15} {:>15} {:>15}".format(eachLine[0], eachLine[1], eachLine[user_choice_1 + 1], eachLine[user_choice_2 + 1]))


# Another way to print (header + list of lists) like a table
# from tabulate import tabulate
# print(tabulate(state_and_region_dataList, headers=[header[0], header[1], header[user_choice_1 + 1], header[user_choice_2 + 1] ]))

```
