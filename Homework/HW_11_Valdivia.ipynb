## Create DataFrame

```python
import pandas as pd

# Use dictionary to create a data
data = {
    'Cities': ['Rome', 'Madrid', 'Istanbul', 'Houston'],
    'Measurement' : [6, 1, 3, 5]
}

# Convert data to a DataFrame
df = pd.DataFrame(data)
df
```

##  Build Data Visualization with Matplotlib

```python
import matplotlib.pyplot as plt

# plt.bar(df['Cities'], df['Measurement'], width = 0.4) #--- Another way to plot in matplotlib
df.plot(kind='bar', x='Cities', y='Measurement', title='City Sizes');
plt.ylabel('Measurement')   # Add title to y-axis
plt.show()
```

## Data Visualization with Seaborn

```python
import seaborn as sns

# Set the figure size
plt.figure(figsize=(7, 5))

# plot a bar chart
sns.barplot(
    y="Measurement", 
    x="Cities", 
    data=df, 
    estimator=sum, 
    ci=None, 
    color='#69b3a2');
```

## Process CSV file

```python
import pandas as pd

df = pd.read_csv('steam.csv')
df
```

## Data Visualization - Pie Chart

```python
# What are the top free-to-play games with better positive ratings

colors = ['#4F6272', '#B7C3F3', '#DD7596', '#8EBA42', '#bc82bd', '#30a2da', '#eeeeee', '#feffb3', '#bcbcbc', '#F0E442'  ]

df[df['price'] == 0].head(10).groupby(
    ['name']
)['positive_ratings'].sum().plot(
    kind='pie',
    figsize=(12,8),
    cmap="Blues_r", # defaults to orangish
    labeldistance = None,
    subplots=True, 
    legend=True,
    colors = colors
)
```
