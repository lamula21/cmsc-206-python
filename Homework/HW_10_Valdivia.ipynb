# Analyzing Files with Pandas

### Process adult.data.txt

```python
# import
import pandas as pd

# Header
header = ['age', 'workclass', 'fnlwgt', 'education','education-num', 'mariscal-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'unknown']

# Read
df = pd.read_csv('adult.data.txt', names=header)

# Delete the last column: 
# del df["unknown"] or 
# df.pop()
df.pop('unknown')
df
```

### People Relationship By Ethnicity

```python
import matplotlib.pyplot as plt

df.groupby(['relationship','race']).size().unstack().plot(kind='bar',stacked=True)

plt.show()
```

### Degree By Country


```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

plt.rcParams.update({'font.size': 20, 'figure.figsize': (30, 18)}) # set font and plot size to be larger

df.groupby(['native-country','education']).size().unstack().plot(kind='bar',stacked=True)

plt.show()
```

### Occupation by Race

```python
import matplotlib.pyplot as plt

df.groupby(['occupation','race']).size().groupby(level=0).apply(
    lambda x: 100 * x / x.sum()
).unstack().plot(kind='bar',stacked=True)

plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter())
plt.show()
```

### Marital Status By Race

```python
import matplotlib.pyplot as plt

df.groupby(['mariscal-status','race']).size().groupby(level=0).apply(
    lambda x: 100 * x / x.sum()
).unstack().plot(kind='bar',stacked=True)

plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter())
plt.show()
```

### Capital Gain by Age

```python
df.plot(kind='scatter', x='age', y='capital-gain', title='Age vs Capital Gain');
```

### Capital Loss by Sex

```python
df.groupby(['capital-loss','sex']).size().groupby(level=0).apply(
    lambda x: 100 * x / x.sum()
).unstack().plot(kind='bar',stacked=True)

plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter())
plt.show()
```

### Hours Per Week By Race

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20, 'figure.figsize': (30, 18)}) # set font and plot size to be larger

df.groupby(['hours-per-week','race']).size().groupby(level=0).apply(
    lambda x: 100 * x / x.sum()
).unstack().plot(kind='bar',stacked=True)


plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter())
plt.show()
```

### Age By Sex

```python
df.groupby(['age','sex']).size().groupby(level=0).apply(
    lambda x: 100 * x / x.sum()
).unstack().plot(kind='bar',stacked=True)

plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter())
plt.show()
```

### Common Group By Age

```python
import matplotlib.pyplot as plt
import pandas as pd

df[['age']].plot(kind='hist',bins=[0,20,40,60,80,100],rwidth=0.8)
plt.show()
```
